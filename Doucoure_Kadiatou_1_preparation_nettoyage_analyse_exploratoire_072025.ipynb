```python
dispo_alim_2017.head()
```

```python
print(dispo_alim_2017['Produit'].unique())
```

```python
df_volaille= dispo_alim_2017[dispo_alim_2017['Produit']=='Viande de Volailles']
df_volaille.head()
```

```python
df_pivot = df_volaille.pivot(index=['Zone','Année'],columns=['Élément'],values='Valeur').fillna(0).reset_index()
colonnes_a_garder = ['Zone','Disponibilité intérieure','Exportations - Quantité','Importations - Quantité','Nourriture','Production']
df_dispo_alim_vf = df_pivot[colonnes_a_garder]
df_dispo_alim_vf.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.2 - Analyse exploiratoire du fichier Population</h3>
</div>

```python
population.head()
```

```python
df_pop_2017_2018= population[population['Année'].isin([2017, 2018])]
df_pop_2017_2018['Année'].unique()
df_pivot_pop= df_pop_2017_2018.pivot(index=['Zone'],columns=['Année'],values='Valeur').fillna(0).reset_index()
df_pivot_pop['Variation_population']=round((df_pivot_pop[2018] - df_pivot_pop[2017]) / df_pivot_pop[2017] * 100,2)
df_pivot_pop_vf=df_pivot_pop[['Zone','Variation_population']]
df_pivot_pop_vf.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.3 - Analyse exploiratoire du fichier PIB</h3>

```python
PIB['Zone'] = PIB['Zone'].apply(lambda x: x.encode('latin1').decode('utf-8') if isinstance(x, str) else x)
PIB.head()
```

```python
PIB_clean = PIB.rename(columns={'AnnÃ©e': 'Annee'})
df_PIB_2017 = PIB_clean[PIB_clean['Annee'] == 2017]
colonnes_a_garder = ['Zone','Valeur']
PIB_2017 = df_PIB_2017[colonnes_a_garder]
PIB_2017.columns=['Zone','PIB']
PIB_2017.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.4 - Analyse exploiratoire du fichier IDH</h3>
</div>

```python
IDH_2017.head()
```

```python
IDH= IDH_2017[IDH_2017['indicator']=='Human Development Index (value)']
df_IDH = IDH.rename(columns={'country': 'Zone','value':'IDH'})
colonnes_a_garder = ['Zone','IDH']
df_IDH = df_IDH[colonnes_a_garder]
df_IDH.columns=['Zone','IDH']
df_IDH.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.5 - Analyse exploiratoire du fichier Instabilité Politique</h3>
</div>

```python
instabilite.head()
```

```python
df_instabilite=instabilite[instabilite['Series Name']=='Political Stability and Absence of Violence/Terrorism: Percentile Rank']
colonne_a_conserver=['Country Name','2017 [YR2017]']
df_instabilite_vf=df_instabilite[colonne_a_conserver]
df_instabilite_vf.columns=['Country Name','2017 [YR2017]']
df_instabilite_vf = df_instabilite_vf.rename(columns={'Country Name': 'Zone','2017 [YR2017]':'Indice stabilite'})
# Insérer la colonne stabilité poilique
df_instabilite_vf['Indice stabilite'] = pd.to_numeric(df_instabilite_vf['Indice stabilite'], errors='coerce')
def classifier_stabilite(valeur):
    if valeur >= 80:
        return "Très stable"
    elif valeur >= 60:
        return "Stable"
    elif valeur >= 40:
        return "Moyenne"
    elif valeur >= 20:
        return "Instable"
    else:
        return "Très instable"
df_instabilite_vf = df_instabilite_vf.copy()
df_instabilite_vf['Stabilité politique'] = df_instabilite_vf['Indice stabilite'].apply(classifier_stabilite)
df_instabilite_vf.head()
```

```python
import pycountry
from babel import Locale
# Définir la locale française
locale_fr = Locale('fr')
# Traduction directe sans fonction
df_instabilite_vf['Zone_fr'] = df_instabilite_vf['Zone'].apply(
    lambda x: locale_fr.territories.get(pycountry.countries.get(name=x).alpha_2) if pycountry.countries.get(name=x) else x
)
```

```python
df_instabilite_vf.head()
```

```python
df_instabilite_final = df_instabilite_vf[['Zone_fr', 'Stabilité politique']]
df_instabilite_final = df_instabilite_final.rename(columns={'Zone_fr': 'Zone'})
df_instabilite_final.head()
```

```python
LPI.head()
```

```python
df_LPI= LPI[['Country','Logistics quality and competence rank', 'Tracking and tracing rank']]
df_LPI = df_LPI.copy()
df_LPI.loc[:,'Zone'] = df_LPI['Country'].apply(
    lambda x: locale_fr.territories.get(pycountry.countries.get(name=x).alpha_2) if pycountry.countries.get(name=x) else x
)
df_LPI.head()
```

```python
df_LPI_vf=df_LPI[['Logistics quality and competence rank','Tracking and tracing rank','Zone']]
df_LPI_vf.rename(columns={"Logistics quality and competence rank":"Rang qualité logistique","Tracking and tracing rank":"Rang trançabilité produits"})
df_LPI_vf.head()
```
